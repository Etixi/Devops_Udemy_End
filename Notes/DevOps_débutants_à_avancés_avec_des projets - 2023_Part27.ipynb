###**289. Présentation**

####**CloudFormation**
+ IAAC (Infra As A Code)

####**Modèles**
+ Fichier texte - YAML ou JSON
+ Modèle de lecture CloudFormation en entrée
+ Décrivez l'état sur l'infra

####**Empiler**
+ Cloudformation a besoin de modèles et crée une pile
+ Nous mettons à jour le modèle et la pile de mise à jour pour apporter des modifications
+ Ensemble de ressources comme une seule unité.

####**Modifier l'ensemble**
+ Avant de mettre à jour une pile, vous pouvez générer un ensemble de modifications.
+ Un ensemble de modifications vous permet de voir comment les modifications auront un impact sur vos ressources existantes.
+ Très important pour les systèmes live.
   + Par exemple, renommer une instance RDS
     + Créer un nouveau
     + Supprimer l'ancien
+ Résultat : Perte de données
